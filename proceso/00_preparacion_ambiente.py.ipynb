```python
# Nombre del Storage Account
storage_account_name = "stfrauddatabricks"

# Containers del proyecto
container_raw = "raw"
container_bronze = "bronze"
container_silver = "silver"
container_gold = "gold"

# Carpeta base del proyecto dentro del Data Lake
project_folder = "fraud_detection"

# COMMAND ----------

# Rutas base en formato ABFSS
raw_base_path = f"abfss://{container_raw}@{storage_account_name}.dfs.core.windows.net/{project_folder}"

bronze_base_path = f"abfss://{container_bronze}@{storage_account_name}.dfs.core.windows.net/{project_folder}"
silver_base_path = f"abfss://{container_silver}@{storage_account_name}.dfs.core.windows.net/{project_folder}"
gold_base_path = f"abfss://{container_gold}@{storage_account_name}.dfs.core.windows.net/{project_folder}"

print("Rutas base configuradas correctamente:")
print("RAW    :", raw_base_path)
print("BRONZE :", bronze_base_path)
print("SILVER :", silver_base_path)
print("GOLD   :", gold_base_path)
```

```python
# Rutas específicas de los datasets en RAW
transactions_raw_path = f"{raw_base_path}/transactions/PS_20174392719_1491204439457_log.csv"
transaction_type_risk_raw_path = f"{raw_base_path}/reference/transaction_type_risk.csv"
customer_profile_raw_path = f"{raw_base_path}/profiles/customer_account_profile.csv"

print("Datasets RAW:")
print("Transacciones        :", transactions_raw_path)
print("Riesgo por tipo      :", transaction_type_risk_raw_path)
print("Perfiles de cuentas  :", customer_profile_raw_path)
```

```python
# Lectura de dataset principal de transacciones
df_transactions_raw = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(transactions_raw_path)
)

display(df_transactions_raw.limit(10))
```

```python
# Lectura de dataset de riesgo por tipo de transacción
df_transaction_type_risk_raw = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(transaction_type_risk_raw_path)
)

display(df_transaction_type_risk_raw)
```

```python
# Lectura de dataset de perfiles de cuentas
df_customer_profile_raw = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(customer_profile_raw_path)
)

display(df_customer_profile_raw.limit(10))
```

```python
# Cantidad de registros por dataset
print("Cantidad de registros por dataset:")

print("Transacciones       :", df_transactions_raw.count())
print("Tipo de riesgo      :", df_transaction_type_risk_raw.count())
print("Perfil de cuentas   :", df_customer_profile_raw.count())
```

```python
# Bloque por esquemas de los datasets
print("Esquema del dataset de transacciones:")
df_transactions_raw.printSchema()

print("Esquema del dataset de riesgo por tipo:")
df_transaction_type_risk_raw.printSchema()

print("Esquema del dataset de perfiles:")
df_customer_profile_raw.printSchema()
```
